```python
from dputils import scrape as sc
import pandas as pd
```

```python
query = 'laptops'
url = f'https://www.amazon.in/s?k={query}'
print(url)
soup = sc.get_webpage_data(url)
```

```python
target ={
    'tag':'div', 'attrs':{'s-main-slot s-result-list s-search-results sg-row'}
}

items = {
    'tag':'div', 'attrs':{'s-result-item'}
}

heading = {
    'tag':'h2', 'attrs':{'a-size-mini a-spacing-none a-color-base s-line-clamp-2'}
}

price = {
    'tag':'span', 'attrs':{'a-price-whole'}
}

delivery_status = {
    'tag':'span', 'attrs':{'a-color-base'}
}

Original_price = {
    'tag':'span', 'attrs':{'a-price a-text-price'}
}
```

```python
output = sc.extract_many(soup, target=target, items=items, heading=heading, price=price, 
delivery_status=delivery_status, Original_price=Original_price)
print(output)
```

```python
df = pd.DataFrame(output)
print(df)
```
