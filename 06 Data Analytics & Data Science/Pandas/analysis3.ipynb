````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
````

````python
print(sns.get_dataset_names())
````

````python
df = sns.load_dataset('titanic')
df.head()
````

# find all first class passenger
- using pandas filter
- using pandas query

````python
filterfirstclass = df['class'] == 'First'
df[filterfirstclass]
````

````python
# filter by age below 20
filterAge20 = df['age'] <= 20 
df[filterAge20]
````

````python
filterage60 = df['age'] >=60
third = df['class'] == 'Third'
df[filterage60 & third]
````

````python
# use pandas query
df.query('age >= 20')
````

````python
# 2nd class females only

filter2ndclassfemales = df['sex'] == 'female'
secondclass = df['class'] == 'Second'
df[filter2ndclassfemales & secondclass]
````

# Group By *Important*
- column to group data
- output columns to display (optional)
- aggreegation function (required)
    - *mean, sum, max, min, std, count*

```
df.groupby('column name')[col_to_out]
```

````python
df.groupby('embark_town').sum()
````

````python
df.groupby('embark_town')['age'].mean()
````

````python
totalfareemale = df['sex'] =='male'


df[totalfareemale]
````

````python
df.groupby('embark_town')['fare'].max()
````

````python
df.groupby('sex')['fare'].sum().plot(title='Fare Male vs Female', kind='bar')
````

````python
df.groupby('embark_town')['survived'].sum()
````

````python
# Multi operation

df.groupby('embark_town')['survived'].agg(['sum', 'count', 'mean'])
````

````python
df.groupby('embark_town')['survived'].agg(['sum', 'count']).plot(title='Survived', kind='barh')
````

````python
df.groupby(['embark_town','who']).sum().plot(title='Survived', kind='barh')
````

````python
#  inline style
df.groupby(['embark_town','who']).sum().style.bar('survived')
````

````python
def age_roundoff(age):
    try:
        return int(age)
    except:
        return 0
````

````python
df.age = df.age.apply(age_roundoff)
````
