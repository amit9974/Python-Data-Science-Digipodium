** Actual processing on this dataset **

````python
df.head(3)
````

````python
cols_to_drop = ['Type','Coverage','AREA','REG','DEV']
df.drop(columns=cols_to_drop, inplace=True)
````

````python
df.head(1)
````

````python
# Rename Coulmn Names
df.rename({
    'OdName':'Country',
    'AreaName': 'Continent',
    'RegName': 'Region',
    'DevName': 'Status'
}, axis=1, inplace=True)
````

````python
# Get status of rename coulmns
df.head(1)
````

````python
df.columns =df.columns.astype(str) # convert column name to string
````

````python
year = list(map(str, range(1980, 2014))) # list of year
````

````python
df[year][:5]   # get the all data of years
````

````python
df['Total'] = df[year].sum(axis=1)  # Sum all data of year wise
 
````

````python
df.set_index('Country', inplace=True)
````

````python
df.iloc[100:]
````

````python
df.loc['Malawi', year].plot(kind='line', figsize=(15,6), color='red', linewidth=3, marker='o')
````
