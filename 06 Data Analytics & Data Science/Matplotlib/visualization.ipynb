```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = np.random.randint(0, 100, 50)
print(data)
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(20,5), facecolor='white')
plt.plot(data, color='red', linewidth=3, 
                linestyle=':', 
                marker='*',
                markersize=10)
```

```python
plt.figure(figsize=(20,5), facecolor='yellow')
plt.plot(data, 'r:*')
plt.title("Graph", fontsize=20, color='red', fontdict={'family':'Times New Roman'})
plt.xlabel("X Axis Graph", fontsize=20, color='blue',fontdict={'family':'Times New Roman'})
plt.ylabel("Y Axis Graph", fontsize= 20, color='blue',fontdict={'family':'Times New Roman'})
```

```python
plt.figure(figsize=(20,5), facecolor='yellow')
plt.plot(data, 'r:*')
plt.title("Graph")
names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']
plt.xticks(np.arange(0, 51, 5),labels=names)
plt.yticks(np.arange(0, 101, 10),labels=names)
plt.savefig("graph1.png")
```

```python
data1 = np.random.randint(0,100,30)
data2 = np.random.randint(0,110,30)
plt.figure(figsize=(20,5), facecolor='white')
plt.plot(data1, label='data1')
plt.plot(data2, label='data2')
plt.legend(loc='upper left', fontsize=20)
plt.show()
```

```python
# Create two diffrent graph
plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.plot(data1, 'r:*', label="data1")
plt.subplot(2,1,2)
plt.plot(data2, 'r:*', label="data2")
plt.legend(loc='upper left')
plt.show()
```
