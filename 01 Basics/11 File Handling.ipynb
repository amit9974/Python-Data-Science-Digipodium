# File handling
In coding we have 3 types of files :

    - text file (.txt, .ini, .md, .html, .css, .js, etc)
    - binary file ( .jpg, .png, .zip, .mp3, .mp4, .avi etc)
    - proprietary file (.docx, .xlsx, .ppt, .pdf, .blend, etc)

we are going to tackle first type

File handling is the process of opening

# Reading a Text File

```python
f = open('test.txt') # load file into a variable
code = f.read() # read the file
f.close() # close the open file
```

```python
print(code)
```

```python
with open('test file handling.txt') as file:  # 2nd way to read the file and its automatic free the storage after use.
    print(file.read())
```

```python
with open('test file handling.txt') as code:
    print(code.read())
```

# writing content in a text

```python
story = "Once upon a time, there was planet that was green"
file = open('story.txt', 'w')
file.write(story)
file.close()
```

```python
print(story)
```

```python
story1 = input('Please enter the story:: ')
with open('story1.txt', 'w') as f:
    f.write(story1)
    f.write('\n')
    f.write('Name - Amit kumar\n')
    f.write('Technical Support Engineer')
    print(story1)
```

# Adding content to a file

```python
data = input('Enter word: ')
with open('new.txt','a') as f:
    f.write("\n"+data)
```
